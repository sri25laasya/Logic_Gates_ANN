**NOR-Gate**

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
import numpy as np
```

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[1],[0],[0], [0]], "float32")
from keras.models import Sequential
from keras.layers.core import Dense
model = Sequential()
model.add(Dense(75, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```python
model.fit(training_data, target_data, epochs=75, verbose=2)
```

```python
model.predict(training_data).round() 
```

```python
model.predict(training_data).round() == target_data
```

```python
model.predict([[0,0]])
model.predict([[1,0]])
model.predict([[0,1]])
model.predict([[1,1]])
```
